应用函数

```python
data=pd.DataFrame(np.random.randn(9,6),columns=list('abcdef'))
```

```python
data
```

```python
data.apply(lambda x:x.max()-x.min(),axis=0)   #每一列最大值-最小值
```

```python
data.a.apply(lambda x: x+10)   #a列应用函数，等于data.a+10, data.a就是一个Series
```

```python
data.applymap(lambda x:x**2+x+3)   #applymap()对每一个元素运算
```

```python
data
```

```python
data['g']=['zl','gll','zr','song','yun','mei','hua','bing','guang']
```

```python
data
```

```python
data.g.apply(lambda x:x.title())   # title() 首字母大写
```

```python
data['g']=data.g.apply(lambda x:x.title())
```

```python
data
```
