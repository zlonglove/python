```python
data1=pd.DataFrame({'one':['a','b','a','a','c','b'],'two':range(6)})
data2=pd.DataFrame({'one':['a','b','c','d'],'two':range(10,14)})
```

```python
data1
```

```python
data2
```

```python
data=pd.merge(data1,data2,on='one') # merge  on,
```

```python
data
```

```python
pd.merge?
```

```python
data=pd.merge(data1,data2,on='one',how='outer') #outer
data
```

```python
data=pd.merge(data1,data2,on='one',how='left') #how='left'data1one
data
```

```python
data=pd.merge(data1,data2,on='one',how='right')
data
```

```python
data1=pd.DataFrame({'one1':['a','b','a','a','c','b'],'two1':range(6)})
data2=pd.DataFrame({'one2':['a','b','c','d'],'two2':range(10,14)})
```

```python
data1
```

```python
data2
```

```python
data=pd.merge(data1,data2,left_on='one1',right_on='one2') #,inner
data
```

```python
data1.index
```

```python
data2.index
```

```python
data=pd.merge(data1,data2,left_index=True,right_index=True,how='outer') #
data
```

```python
data=pd.merge(data1,data2,left_on='two1',right_index=True,how='outer')
data
```

```python
data=pd.merge(data1,data2,left_on='one1',right_index=True,how='outer')
data
```
