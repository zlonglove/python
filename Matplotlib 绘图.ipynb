```python
x=[1,3,5,7]
y=[2,6,3,9]
```

```python
plt.barh(x,y)  #横向条形图
```

```python
y1=[2,4,6,1]
```

堆积图、饼图

```python
com=[1,2,3,4] #年份  各个公司市场份额变化
y1=[24,27,35,43]
y2=[30,35,17,12]
y3=[27,13,15,8]
y4=[19,25,33,37]
```

```python
labels=['apple','micro','yahoo','google']
```

```python
plt.stackplot(com,y1,y2,y3,y4,labels=labels)
plt.legend(loc=1)
plt.xticks([1,2,3,4],['2011','2012','2013','2014'])
plt.title('market')
```

```python
labels=['apple','micro','yahoo','google']
size=[15,30,45,10]   #占有率
```

```python
explode=(0.1,0.1,0,0)  #将第一个第二个label突出，数值越大，越突出
```

```python
plt.pie(size,labels=labels,autopct='%1.1f%%',explode=explode)       # .1小数点后面1位，float,%号
plt.show()
```

```python
plt.figure(figsize=(8,8))
plt.pie(size,labels=labels,autopct='%1.1f%%',explode=explode,startangle=90,shadow=True)  #控制位置,加阴影    
plt.show()
```

绘制多图

```python
x=[1,2,3,4]
y1=[4,5,8,9]
y2=[2,5,3,8]
```

```python
plt.subplot(2,1,1)  #2行1列的第一个
plt.plot(x,y1)
plt.subplot(2,1,2)
plt.plot(x,y2)
plt.suptitle('y1&y2')  #总title
```

```python
plt.subplot(2,3,1)
plt.plot(x,y1)
plt.subplot(2,3,2)
plt.bar(x,y1)
plt.subplot(2,3,3)
plt.barh(x,y1)
plt.subplot(2,3,4)
plt.scatter(x,y1)
plt.subplot(2,3,5)
plt.boxplot(x,y1)
plt.show()
```

```python
plt.subplot(2,3,1)
plt.plot(x,y1)
plt.subplot(2,3,2)
plt.bar(x,y1)
plt.subplot(2,3,3)
plt.barh(x,y1)
plt.subplot(2,1,2)  #最下面相当于2行1列第二个
plt.plot(x,y2,'b--')
```

面向对象绘图

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data=np.random.randn(1000)
```

```python
fig,ax=plt.subplots()  #初始化图像及坐标系
ax.hist(data,bins=20)
```

```python
fig,ax=plt.subplots(2,2)
```

```python
ax   #2行2列的array
```

```python
fig,ax=plt.subplots(2,2,figsize=(8,8))
ax[0][0].hist(data,label='hist')  #切片，在第一行第一列绘图
ax[0][1].boxplot(data)
plt.tight_layout()     #让多个图形显示的不拥挤
```

```python
fig,ax=plt.subplots() #初始化图像及坐标系
ax.hist(data,bins=50,label='here')
ax.set_title('data')
ax.set_xlabel('x')
ax.legend(loc=2)
ax.set_xticks([-3,0,3])
ax.set_xticklabels(['low','median','high'])
ax.grid(True)
```
